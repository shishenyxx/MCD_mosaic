```r
library(maftools)
```

```r
laml = read.maf(maf = 'oncoplot.maf',clinicalData = 'ILEA_classification.tsv',verbose = TRUE)
```

```r
laml
```

```r
options(repr.plot.width=10, repr.plot.height=15)
oncoplot(maf = laml, clinicalFeatures = 'ILEA_classification', sortByAnnotation = TRUE ,top = 100, draw_titv = TRUE)
```

```r
pdf("FCD_oncoplot.pdf", width = 10, height = 20, useDingbats = F)
oncoplot(maf = laml, clinicalFeatures = 'ILEA_classification', sortByAnnotation = TRUE ,top = 100, draw_titv = TRUE)
dev.off()
```

```r
options(repr.plot.width=20, repr.plot.height=15)
#Shows sample summry.
getSampleSummary(laml)
#Shows gene summary.
getGeneSummary(laml)
#shows clinical data associated with samples
getClinicalData(laml)
#Shows all fields in MAF
getFields(laml)
#Writes maf summary to an output file with basename laml.
write.mafSummary(maf = laml, basename = 'laml')
plotmafSummary(maf = laml, addStat = 'median')
```

```r
#lollipop plot for DNMT3A, which is one of the most frequent mutated gene in Leukemia.
lollipopPlot(
  maf = laml,
  gene = 'MTOR',
  AACol = 'Protein_Change',
  showMutationRate = TRUE,
  labelPos = 882
)
```
